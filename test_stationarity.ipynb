```python
from dlforecasting import *
from dlforecasting.generate_series import rw_generator
import matplotlib.pyplot as plt
import numpy as np
```

```python
rw_gen = rw_generator(1000, baseline=10)
ts = rw_gen.generate()
plt.plot(ts)
```

```python
from dlforecasting.stationary import Stationarity
sta = Stationarity()
diff = sta.fit_transform(ts)
plt.plot(diff)
```

```python
dediff = sta.dediff(diff)
plt.plot(dediff)
```
